```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from scipy import stats
import matplotlib.ticker as ticker
```

```python
N_al = np.array([124112, 73533, 45043, 27879, 17773, 11589, 7677])
er_al = np.array([733, 778, 381, 402, 154,152, 107])
th_al = 2
num_al = np.array([1,2,3,4,5,6,7]) * th_al 
```

```python
er_al/N_al *100
```

```python
N_pb = np.array([107877, 53225, 27964, 15559, 8513, 4535, 2592])
er_pb = np.array([475, 578, 251, 193, 123, 94, 54])
th_pb = 0.5
num_pb = np.array([1,2,3,4,5,6,7]) * th_pb 
```

```python
er_pb/N_pb *100
```

```python
N_fe = np.array([106262, 52335, 27538, 14752, 8090, 4386, 2399])
er_fe = np.array([1014, 576, 419, 137, 125, 67, 44])
th_fe = 1
num_fe = np.array([1,2,3,4,5,6,7]) * th_fe
```

```python
er_fe/N_fe *100
```

```python
m_fe = 7.874 * th_fe
m_pb = 11.3415 * th_pb
m_al = 2.7 * th_al
```

```python
def approx(x, a, b):
    return a * x + b
beta_opt1, beta_cov1 = curve_fit(approx, num_al, np.log(N_al), sigma = er_al/N_al)
a = beta_opt1[0]
b = beta_opt1[1]


print('a = ', a)
print('b = ', b)



#получим погрешности для коэффициентов
sigma_a = np.sqrt(beta_cov1[0,0])
sigma_b = np.sqrt(beta_cov1[1,1])


print('sigma_a = ', sigma_a)
print('sigma_b = ', sigma_b)


residuals1 = np.log(N_al) - approx(num_al,*beta_opt1)
fres1 = sum(residuals1**2)
S_e = np.sqrt(fres1/len(num_al))
print('S_e = ', S_e)


print('Relative S_r a = % ', 100* sigma_a/abs(a))
print('Relative S_r b = % ', 100* sigma_b/abs(b))



# график
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(num_al, approx(num_al, *beta_opt1), 'coral', lw = 2)
ax.scatter(num_al, np.log(N_al), marker = '*', color = 'red', s = 10)
lgnd = ax.legend([r'Aluminum fit $\ln{N}(l)$ = %.3f $\cdot$ $l$ + %.3f ' %(a,b), 'Aluminum Data'], loc='upper right', shadow=True)

# ax.set_title(r'График Ферми')
ax.set_ylabel(r'$\ln{N}$')
ax.set_xlabel(r'$l$, см')

ax.grid(which='major',
color = 'k')

ax.minorticks_on()

ax.grid(which='minor',
color = 'gray',
linestyle = ':', linewidth = 0.5)

ax.grid(which='major', linewidth = 0.5)
plt.errorbar(num_al, np.log(N_al), fmt = 'ro', marker = '*', markersize = '4', yerr = S_e, capsize = 4, elinewidth = 2, capthick = 2, ecolor = 'violet')


# plt.xlim (-10, 850)
# plt.ylim (-10, 245)
plt.savefig('Al.png', dpi=400, quality=100)
plt.show()
```

```python
mu_al1 = a*th_al/m_al
mu_al1
```
